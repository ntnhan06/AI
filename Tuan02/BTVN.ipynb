```python
#Bai01
def tron_mang(a, b):
    c = []
    min_len = min(len(a), len(b))

    for i in range(min_len):
        c.append(min(a[i], b[i]))

    if len(a) > len(b):
        c.extend(a[min_len:])
    else:
        c.extend(b[min_len:])

    return c

a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

c = tron_mang(a, b)
print("Mang ket qua:", c)
```

```python
#Bai02
m = int(input("Nhap so dong m: "))
n = int(input("Nhap so cot n: "))

a = []
print("Nhap ma tran:")
for i in range(m):
    row = list(map(int, input().split()))
    a.append(row)

if m == n:
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    print("\nTong tam giac tren (cheo phu):", tong)

for i in range(m):
    for j in range(n):
        if a[i][j] < 0:
            a[i][j] = abs(a[i][j])

print("\nMa tran sau khi doi so am:")
for row in a:
    print(*row)

x = int(input("\nNhap x: "))
for i in range(m):
    for j in range(n):
        if a[i][j] % 2 == 0:
            a[i][j] = x

print("\nMa tran sau khi thay so chan:")
for row in a:
    print(*row)

toan_chan = all(a[i][j] % 2 == 0 for i in range(m) for j in range(n))
print("\nMa tran toan chan?", toan_chan)

if m == n:
    doi_xung = True
    for i in range(n):
        for j in range(n):
            if a[i][j] != a[j][i]:
                doi_xung = False
                break
    print("Ma tran doi xung?", doi_xung)

    tang = True
    for i in range(n - 1):
        if a[i][i] > a[i+1][i+1]:
            tang = False
            break
    print("Cheo chinh tang dan?", tang)

    print("\nTam giac duoi (cheo phu):")
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(a[i][j], end=" ")
        print()

    giam = True
    for i in range(n - 1):
        if a[i][n-1-i] < a[i+1][n-2-i]:
            giam = False
            break
    print("Cheo phu giam dan?", giam)
```

```python
#Bai03
data = {}

for i in range(1, 4):
    print(f"\nNhap sinh vien {i}:")
    ma = input("Ma: ")
    ten = input("Ten: ")
    diem = float(input("Diem: "))

    data[f"sv{i}"] = {
        "ma": ma,
        "ten": ten,
        "diem": diem
    }

print("\nDanh sach sinh vien:")
for key, value in data.items():
    print(key, ":", value)

print("\nTruy cap du lieu:")
print("Ten sv1:", data["sv1"]["ten"])
print("Diem sv3:", data["sv3"]["diem"])
```

```python
#Bai05
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "MaSV": ["001", "002", "003", "004"],
    "Ten": ["An", "Binh", "Lan", "Phan"],
    "Diem": [7.5, 8.0, 9.0, 6.5],
    "Tuoi": [20, 21, 19, 20]
}

df = pd.DataFrame(data)
print("=== Danh sach sinh vien ===")
print(df)

print("\n=== Sinh vien diem >= 8 ===")
print(df[df["Diem"] >= 8])

diem_tb = df["Diem"].mean()
print(f"\nDiem TB cua lop: {diem_tb:.2f}")

plt.bar(df["Ten"], df["Diem"], color="skyblue")
plt.title("Diem sinh vien")
plt.xlabel("Ten")
plt.ylabel("Diem")
plt.axhline(y=diem_tb, color='r', linestyle='--', label=f"TB lop: {diem_tb:.2f}")
plt.legend()
plt.show()

plt.plot(df["Ten"], df["Tuoi"], marker="o", color="green")
plt.title("Tuoi sinh vien")
plt.xlabel("Ten")
plt.ylabel("Tuoi")
plt.show()
```
