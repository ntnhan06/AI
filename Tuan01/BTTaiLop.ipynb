```python
#Bai13
def tron_mang(a, b):
    kq = []
    n = min(len(a), len(b))

    # cộng phần tử tương ứng
    for i in range(n):
        kq.append(a[i] + b[i])

    # phần dư của a
    for i in range(n, len(a)):
        kq.append(a[i])

    # phần dư của b
    for i in range(n, len(b)):
        kq.append(b[i])

    return kq


a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

print(tron_mang(a, b))
```

```python
#bai14
#a. Tạo mảng 2 chiều chứa các số nguyên ngẫu nhiên.
import random
import math

def create_random_matrix(m, n, min_val=0, max_val=100):
    """
    Tạo một mảng 2 chiều (ma trận) với m dòng và n cột chứa các số nguyên ngẫu nhiên.
    """
    matrix = []
    for _ in range(m):
        row = [random.randint(min_val, max_val) for _ in range(n)]
        matrix.append(row)
    return matrix

m = int(input("Nhập số dòng (m): "))
n = int(input("Nhập số cột (n): "))

a = create_random_matrix(m, n)
print("Mảng a:")
for row in a:
    print(row)
#b. Xuất các phần tử thuộc dòng k.
def print_row_k(matrix, k):
    """
    Xuất các phần tử thuộc dòng k của ma trận.
    Dòng được đánh số từ 0.
    """
    if 0 <= k < len(matrix):
        print(f"Các phần tử thuộc dòng {k}: {matrix[k]}")
    else:
        print(f"Dòng {k} không hợp lệ. Ma trận có {len(matrix)} dòng (0 đến {len(matrix)-1}).")

k_row = int(input("Nhập dòng k muốn xuất (0-based index): "))
print_row_k(a, k_row)
#c. Xuất các phần tử thuộc cột k
def print_column_k(matrix, k):
    """
    Xuất các phần tử thuộc cột k của ma trận.
    Cột được đánh số từ 0.
    """
    if not matrix:
        print("Ma trận rỗng.")
        return

    num_cols = len(matrix[0])
    if 0 <= k < num_cols:
        column_elements = [row[k] for row in matrix]
        print(f"Các phần tử thuộc cột {k}: {column_elements}")
    else:
        print(f"Cột {k} không hợp lệ. Ma trận có {num_cols} cột (0 đến {num_cols-1}).")

k_col = int(input("Nhập cột k muốn xuất (0-based index): "))
print_column_k(a, k_col)
#d. Tìm dòng tổng lớn nhất
def find_row_with_max_sum(matrix):
    if not matrix:
        return -1, 0

    max_sum = -float('inf')
    max_sum_row_index = -1

    for i, row in enumerate(matrix):
        current_sum = sum(row)
        if current_sum > max_sum:
            max_sum = current_sum
            max_sum_row_index = i

    return max_sum_row_index, max_sum

max_sum_row_index, max_sum_value = find_row_with_max_sum(a)
if max_sum_row_index != -1:
    print(f"Dòng có tổng lớn nhất là dòng {max_sum_row_index} với tổng = {max_sum_value}")
    print(f"Các phần tử của dòng đó: {a[max_sum_row_index]}")
else:
    print("Ma trận rỗng, không tìm thấy dòng có tổng lớn nhất.")
#e. Tìm dòng có tích nhỏ nhất

def find_column_with_min_product(matrix):
    """
    Tìm cột có tích các phần tử nhỏ nhất trong ma trận.
    Trả về chỉ số cột và tích nhỏ nhất.
    """
    if not matrix or not matrix[0]:
        return -1, 0

    num_rows = len(matrix)
    num_cols = len(matrix[0])

    min_product = float('inf')
    min_product_col_index = -1

    for j in range(num_cols):
        current_product = 1
        for i in range(num_rows):
            current_product *= matrix[i][j]

        if current_product < min_product:
            min_product = current_product
            min_product_col_index = j

    return min_product_col_index, min_product

min_product_col_index, min_product_value = find_column_with_min_product(a)
if min_product_col_index != -1:
    print(f"Cột có tích nhỏ nhất là cột {min_product_col_index} với tích = {min_product_value}")
    column_elements = [row[min_product_col_index] for row in a]
    print(f"Các phần tử của cột đó: {column_elements}")
else:
    print("Ma trận rỗng hoặc không có cột, không tìm thấy cột có tích nhỏ nhất.")
#f.
def print_even_rows_odd_cols(matrix):
    """
    Xuất các phần tử thuộc dòng chẵn và cột lẻ trong ma trận.
    """
    if not matrix or not matrix[0]:
        print("Ma trận rỗng.")
        return

    num_rows = len(matrix)
    num_cols = len(matrix[0])
    result_elements = []

    print("Các phần tử thuộc dòng chẵn và cột lẻ:")
    for i in range(num_rows):
        if i % 2 == 0:  # Dòng chẵn (0, 2, 4,...)
            for j in range(num_cols):
                if j % 2 != 0:  # Cột lẻ (1, 3, 5,...)
                    result_elements.append(matrix[i][j])
                    print(f"a[{i}][{j}] = {matrix[i][j]}")
    if not result_elements:
        print("Không có phần tử nào thỏa mãn điều kiện.")

print_even_rows_odd_cols(a)
#g. Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của a.
def average_even_elements_in_odd_rows(matrix):
    """
    Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của ma trận.
    """
    if not matrix or not matrix[0]:
        return 0

    sum_even_elements = 0
    count_even_elements = 0

    num_rows = len(matrix)

    for i in range(num_rows):
        if i % 2 != 0:  # Dòng lẻ (1, 3, 5,...)
            for element in matrix[i]:
                if element % 2 == 0:  # Phần tử chẵn
                    sum_even_elements += element
                    count_even_elements += 1

    if count_even_elements > 0:
        return sum_even_elements / count_even_elements
    else:
        return 0

avg = average_even_elements_in_odd_rows(a)
if avg > 0:
    print(f"Trung bình cộng các phần tử chẵn thuộc dòng lẻ: {avg:.2f}")
else:
    print("Không có phần tử chẵn nào thuộc dòng lẻ để tính trung bình cộng.")
#h. Tính trung bình cộng các phần tử thuộc biên.
def average_border_elements(matrix):
    """
    Tính trung bình cộng các phần tử thuộc biên của ma trận.
    """
    if not matrix or not matrix[0]:
        return 0

    num_rows = len(matrix)
    num_cols = len(matrix[0])

    if num_rows == 1 and num_cols == 1:
        return matrix[0][0]

    border_elements = []


    for j in range(num_cols):
        border_elements.append(matrix[0][j])

    if num_rows > 1:
        for j in range(num_cols):
            border_elements.append(matrix[num_rows - 1][j])

    for i in range(1, num_rows - 1):
        border_elements.append(matrix[i][0])

    if num_cols > 1:
        for i in range(1, num_rows - 1):
            border_elements.append(matrix[i][num_cols - 1])

    unique_border_elements = list(set(border_elements))

    if unique_border_elements:
        return sum(unique_border_elements) / len(unique_border_elements)
    else:
        return 0

avg_border = average_border_elements(a)
if a and a[0]:
    print(f"Trung bình cộng các phần tử thuộc biên: {avg_border:.2f}")
else:
    print("Ma trận rỗng, không có phần tử biên nào.")
```

```python
#Bai15
from datetime import datetime

class SinhVien:
    def __init__(self, ma, ten, namsinh, diem):
        self.ma = ma
        self.ten = ten
        self.namsinh = namsinh
        self.diem = diem

n = int(input("Nhap so sinh vien: "))
ds = []

for i in range(n):
    print(f"\nNhap sinh vien thu {i+1}:")
    ma = input("Ma SV: ")
    ten = input("Ten: ")
    namsinh = int(input("Nam sinh: "))
    diem = float(input("Diem TB: "))
    ds.append(SinhVien(ma, ten, namsinh, diem))

dem_lenlop = sum(1 for sv in ds if sv.diem >= 5)
print("\nSo sinh vien du dieu kien len lop:", dem_lenlop)

nam_hien_tai = datetime.now().year
print("\nSinh vien du 20 tuoi:")
for sv in ds:
    if nam_hien_tai - sv.namsinh >= 20:
        print(sv.ma, sv.ten, sv.namsinh, sv.diem)

dem_dh = sum(1 for sv in ds if len(sv.ma) >= 4 and sv.ma[2:4] == "DH")
print("\nSo sinh vien he DH:", dem_dh)

dem_lan = sum(1 for sv in ds if sv.ten.split()[-1] == "Lan")
print("\nSo sinh vien ten Lan:", dem_lan)

dem_phan = sum(1 for sv in ds if sv.ten.split()[0] == "Phan")
print("\nSo sinh vien ho Phan:", dem_phan)
```
