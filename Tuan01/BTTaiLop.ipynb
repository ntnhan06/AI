```python
#Bai13
def tron_mang(a, b):
    kq = []
    n = min(len(a), len(b))

    # cộng phần tử tương ứng
    for i in range(n):
        kq.append(a[i] + b[i])

    # phần dư của a
    for i in range(n, len(a)):
        kq.append(a[i])

    # phần dư của b
    for i in range(n, len(b)):
        kq.append(b[i])

    return kq


a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

print(tron_mang(a, b))
```

```python
#Bai14
import random

m = int(input("Nhap so dong m: "))
n = int(input("Nhap so cot n: "))

a = [[random.randint(0, 99) for _ in range(n)] for _ in range(m)]

print("\nMa tran:")
for row in a:
    print(*row)

k = int(input("\nNhap dong k: "))
print("Dong", k, ":", *a[k])

k = int(input("\nNhap cot k: "))
print("Cot", k, ":")
for i in range(m):
    print(a[i][k])

max_sum = -1
index = -1
for i in range(m):
    s = sum(a[i])
    if s < 45 and s > max_sum:
        max_sum = s
        index = i
print("\nDong co tong <45 lon nhat:", index)

min_product = None
min_col = 0

for j in range(n):
    product = 1
    for i in range(m):
        product *= a[i][j]
    if min_product is None or product < min_product:
        min_product = product
        min_col = j

print("Cot co tich nho nhat:", min_col)

print("\nPhan tu dong chan, cot le:")
for i in range(0, m, 2):
    for j in range(1, n, 2):
        print(a[i][j], end=" ")
    print()

s = 0
count = 0
for i in range(1, m, 2):
    for j in range(n):
        if a[i][j] % 2 == 0:
            s += a[i][j]
            count += 1

if count > 0:
    print("TBC chan dong le:", s / count)
else:
    print("Khong co gia tri hop le")

s = 0
count = 0
for i in range(m):
    for j in range(n):
        if i == 0 or i == m-1 or j == 0 or j == n-1:
            s += a[i][j]
            count += 1

print("TBC bien:", s / count)

product = 1
count = 0
for i in range(1, m-1):
    for j in range(1, n-1):
        product *= a[i][j]
        count += 1

if count > 0:
    print("TBC tich khong bien:", product / count)
else:
    print("Khong co phan tu ben trong")
```

```python
#Bai15
from datetime import datetime

class SinhVien:
    def __init__(self, ma, ten, namsinh, diem):
        self.ma = ma
        self.ten = ten
        self.namsinh = namsinh
        self.diem = diem

n = int(input("Nhap so sinh vien: "))
ds = []

for i in range(n):
    print(f"\nNhap sinh vien thu {i+1}:")
    ma = input("Ma SV: ")
    ten = input("Ten: ")
    namsinh = int(input("Nam sinh: "))
    diem = float(input("Diem TB: "))
    ds.append(SinhVien(ma, ten, namsinh, diem))

dem_lenlop = sum(1 for sv in ds if sv.diem >= 5)
print("\nSo sinh vien du dieu kien len lop:", dem_lenlop)

nam_hien_tai = datetime.now().year
print("\nSinh vien du 20 tuoi:")
for sv in ds:
    if nam_hien_tai - sv.namsinh >= 20:
        print(sv.ma, sv.ten, sv.namsinh, sv.diem)

dem_dh = sum(1 for sv in ds if len(sv.ma) >= 4 and sv.ma[2:4] == "DH")
print("\nSo sinh vien he DH:", dem_dh)

dem_lan = sum(1 for sv in ds if sv.ten.split()[-1] == "Lan")
print("\nSo sinh vien ten Lan:", dem_lan)

dem_phan = sum(1 for sv in ds if sv.ten.split()[0] == "Phan")
print("\nSo sinh vien ho Phan:", dem_phan)
```
