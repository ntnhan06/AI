```python
#Bai11
n = int(input("Nhập số phần tử: "))

a = []
for i in range(n):
    x = int(input(f"Nhập phần tử a[{i}]: "))
    a.append(x)

b = []
for x in a:
    if x % 2 == 0:   # số chẵn
        b.append(x)

print("Mảng a:", a)
print("Mảng b (chẵn):", b)
```

```python
#Bai13
def tron_mang(a, b):
    kq = []
    n = min(len(a), len(b))

    # cộng phần tử tương ứng
    for i in range(n):
        kq.append(a[i] + b[i])

    # phần dư của a
    for i in range(n, len(a)):
        kq.append(a[i])

    # phần dư của b
    for i in range(n, len(b)):
        kq.append(b[i])

    return kq


a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

print(tron_mang(a, b))
```

```python
#Bai14
import random

def tao_mang(m, n):
    a = []
    for i in range(m):
        row = []
        for j in range(n):
            row.append(random.randint(0, 100))
        a.append(row)
    return a

def xuat_dong(a, k):
    print(a[k])

def xuat_cot(a, k):
    for i in range(len(a)):
        print(a[i][k], end=" ")
    print()

def dong_tong_max(a):
    max_sum = sum(a[0])
    dong = 0

    for i in range(len(a)):
        s = sum(a[i])
        if s > max_sum:
            max_sum = s
            dong = i

    return dong
```
