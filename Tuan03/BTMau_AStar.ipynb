```python
#BTMau A*
from collections import deque

class Graph:
    def __init__(self, adjac_lis):
      self.adjac_lis = adjac_lis

    def get_neighbors(self, v):
      return self.adjac_lis[v]

    def h(self, n):
      H={
          'A':1,
          'B':1,
          'C':1,
          'D':1
      }
      return H[n]

    def a_star_algorithm(self, start, stop):
      open_lst = set([start])
      closed_lst = set() # Use empty set literal

      poo = {}
      poo[start] = 0

      par = {}
      par[start] = start

      while len(open_lst) > 0:
        n = None

        # Find the node with the lowest f(n) = g(n) + h(n) in open_lst
        for v in open_lst:
          if n is None or (poo[v] + self.h(v)) < (poo[n] + self.h(n)):
            n = v

        if n is None:
          print('Path does not exist?')
          return None

        if n == stop:
          reconst_path = []

          while par[n] != n:
            reconst_path.append(n)
            n = par[n]
          reconst_path.append(start)
          reconst_path.reverse()

          print('Path found: {}'.format(reconst_path))
          return reconst_path

        # Remove n from the open_lst and add to closed_lst
        open_lst.remove(n)
        closed_lst.add(n)

        # Process neighbors of n
        for (m, weigh) in self.get_neighbors(n):
          # If the neighbor is not in open_lst and not in closed_lst
          if m not in open_lst and m not in closed_lst:
            open_lst.add(m)
            par[m] = n
            poo[m] = poo[n] + weigh
          # If the neighbor is already in open_lst or closed_lst, check if the new path to m is shorter
          else:
            if poo[m] > poo[n] + weigh:
              poo[m] = poo[n] + weigh
              par[m] = n

              # If m was in closed_lst, move it back to open_lst because we found a better path to it
              if m in closed_lst:
                closed_lst.remove(m)
                open_lst.add(m)

      print('Path does not exist!')
      return None

# Example usage of the Graph and A* algorithm
adjac_lis = {
    'A':[('B', 1), ('C', 3), ('D', 7)],
    'B':[('C', 1), ('D', 5)],
    'C':[('D',12)],
    'D':[] # D has no neighbors
}

graph1 = Graph(adjac_lis)
path = graph1.a_star_algorithm('A','C')
print(path)

```
