```python
#Bai01
import math

# Kiểm tra người thắng
def check_winner(board):
    # Hàng
    for row in board:
        if row[0] == row[1] == row[2] != '_':
            return row[0]

    # Cột
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] != '_':
            return board[0][col]

    # Đường chéo
    if board[0][0] == board[1][1] == board[2][2] != '_':
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != '_':
        return board[0][2]

    return None

# Kiểm tra còn nước đi không
def is_moves_left(board):
    for row in board:
        if '_' in row:
            return True
    return False

# Hàm minimax
def minimax(board, depth, is_max):
    winner = check_winner(board)

    # Trạng thái kết thúc
    if winner == 'X':
        return 1
    elif winner == 'O':
        return -1
    elif not is_moves_left(board):
        return 0

    if is_max:  # X chơi
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = 'X'
                    best = max(best, minimax(board, depth + 1, False))
                    board[i][j] = '_'
        return best
    else:  # O chơi
        best = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = 'O'
                    best = min(best, minimax(board, depth + 1, True))
                    board[i][j] = '_'
        return best

# Tìm nước đi tốt nhất cho X
def find_best_move(board):
    best_val = -math.inf
    best_move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if board[i][j] == '_':
                board[i][j] = 'X'
                move_val = minimax(board, 0, False)
                board[i][j] = '_'

                if move_val > best_val:
                    best_move = (i, j)
                    best_val = move_val

    return best_move

# Ví dụ sử dụng
board = [
    ['X', 'O', 'X'],
    ['O', 'O', '_'],
    ['_', 'X', '_']
]

best_move = find_best_move(board)
print("Nước đi tốt nhất cho X:", best_move)
```

```python
#Bai02
import math

# Kiểm tra người thắng
def check_winner(board):
    # Hàng
    for row in board:
        if row[0] == row[1] == row[2] != '_':
            return row[0]

    # Cột
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] != '_':
            return board[0][col]

    # Đường chéo
    if board[0][0] == board[1][1] == board[2][2] != '_':
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != '_':
        return board[0][2]

    return None

# Kiểm tra còn nước đi không
def is_moves_left(board):
    for row in board:
        if '_' in row:
            return True
    return False

# Hàm alpha-beta
def alphabeta(board, depth, alpha, beta, is_max):
    winner = check_winner(board)

    # Trạng thái kết thúc
    if winner == 'X':
        return 10 - depth
    elif winner == 'O':
        return depth - 10
    elif not is_moves_left(board):
        return 0

    if is_max:  # X (MAX)
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = 'X'
                    val = alphabeta(board, depth + 1, alpha, beta, False)
                    board[i][j] = '_'

                    best = max(best, val)
                    alpha = max(alpha, best)

                    # Cắt tỉa
                    if beta <= alpha:
                        return best
        return best
    else:  # O (MIN)
        best = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = 'O'
                    val = alphabeta(board, depth + 1, alpha, beta, True)
                    board[i][j] = '_'

                    best = min(best, val)
                    beta = min(beta, best)

                    # Cắt tỉa
                    if beta <= alpha:
                        return best
        return best

# Tìm nước đi tốt nhất cho X
def find_best_move(board):
    best_val = -math.inf
    best_move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if board[i][j] == '_':
                board[i][j] = 'X'
                move_val = alphabeta(board, 0, -math.inf, math.inf, False)
                board[i][j] = '_'

                if move_val > best_val:
                    best_val = move_val
                    best_move = (i, j)

    return best_move

# Ví dụ sử dụng
board = [
    ['X', 'O', 'X'],
    ['O', 'O', '_'],
    ['_', 'X', '_']
]

best_move = find_best_move(board)
print("Nước đi tốt nhất cho X:", best_move)
```
